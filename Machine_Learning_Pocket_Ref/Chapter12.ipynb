### Create Features

```python
name = df.name
```

```python
df = df.drop(columns=["name",
                      "ticket",
                      "home.dest",
                      "boat",
                      "body",
                      "cabin"])
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df['sex_male'] = df['sex_male'].astype('int64')
df['embarked_Q'] = df['embarked_Q'].astype('int64')
df['embarked_S'] = df['embarked_S'].astype('int64')
```

```python
df.columns
```

### Split Data

```python
y = df.survived
X = df.drop(["survived"], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.isnull().sum()
```

### Impute Data

```python
# To be used for missing value imputation
from sklearn.impute import SimpleImputer
```

```python
# Fix null values in X dataframe for Chapter 8
imp_mode = SimpleImputer(missing_values = np.nan, strategy="median")
cols_to_impute = ["age", "fare"]
X[cols_to_impute] = imp_mode.fit_transform(X[cols_to_impute])
```

```python
imp_mode = SimpleImputer(missing_values = np.nan, strategy="median")
cols_to_impute = ["age", "fare"]
X_train[cols_to_impute] = imp_mode.fit_transform(X_train[cols_to_impute])
```

```python
X_test.isnull().sum()
```

```python
imp_mode = SimpleImputer(missing_values = np.nan, strategy="median")
cols_to_impute = ["age"]
X_test[cols_to_impute] = imp_mode.fit_transform(X_test[cols_to_impute])
```

```python
X_test.isnull().sum()
```

### Normalize Data

```python
# To be used for data scaling (normalizing)
from sklearn.preprocessing import StandardScaler
```

```python
# easier way of building out a list when you have a lot of values/columns to add 
cols = "pclass,age,sibsp,parch,fare".split(",")
#cols = X_train.columns
cols
```

```python
sca = StandardScaler()
sca.fit(X_train[cols])
X_train[cols] = sca.transform(X_train[cols])

sca.fit(X_test[cols])
X_test[cols] = sca.transform(X_test[cols])
```

### Chapter 12 begins here

```python
pd.set_option('display.float_format', '{:.10f}'.format)
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42, max_depth=3)
dt.fit(X_train, y_train)
```

### Confusion Matrix

```python
# create confusion matrix using pandas
y_predict = dt.predict(X_test)
tp = ((y_test == 1) & (y_test == y_predict)).sum()
tn = ((y_test == 0) & (y_test == y_predict)).sum()
fp = ((y_test == 0) & (y_test != y_predict)).sum()
fn = ((y_test == 1) & (y_test != y_predict)).sum()

print (f"True Negative:{tn}\n"
       f"False Positive:{fp}\n"
       f"False Negative:{fn}\n"
       f"True Positive:{tp}")
       
```

```python
# create confusion matrix using sklearn
from sklearn.metrics import confusion_matrix

y_predict = dt.predict(X_test)
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=["Predict Died","Predict Survive"],
    index=["True Death","True Survive"])
```

```python
# create confusion matrix using yellowbrick
from yellowbrick.classifier import ConfusionMatrix

mapping = {0:"Died",1:"Survived"}
fig, ax = plt.subplots(figsize=(6,6))
cm_viz = ConfusionMatrix(
    dt,
    classes=["Died","Survived"],
    label_encoder=mapping)
cm_viz.score(X_test, y_test)
cm_viz.poof()
```

### Accuracy

```python
(tp + tn) / (tp+tn+fp+fn)
```

```python
from sklearn.metrics import accuracy_score
y_predict=dt.predict(X_test)
accuracy_score(y_test, y_predict)
```

### Recall

```python
tp / (tp+fn)
```

```python
from sklearn.metrics import recall_score
y_predict=dt.predict(X_test)
recall_score(y_test, y_predict)
```

### Precision

```python
tp / (tp+fp)
```

```python
from sklearn.metrics import precision_score
y_predict=dt.predict(X_test)
precision_score(y_test,y_predict)
```

### F1

```python
pre = tp/(tp+fp)
rec = tp/(tp+fn)
(2*pre*rec)/(pre+rec)
```

```python
from sklearn.metrics import f1_score
y_predict=dt.predict(X_test)
f1_score(y_test, y_predict)
```
